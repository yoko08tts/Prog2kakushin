```python
import numpy as np
import IPython

notes = ['', '#', '', '', '#', '', '#', '', '', '#', '', '#', '']

tone_range = '\u9AD8\u97F3\u57DF' #@param ['', '', ''] {allow-input: false}

if tone_range == '':
  bass = 110
elif tone_range == '':
  bass = 220
elif tone_range == '':
  bass = 440

f = list(bass * 2 ** (np.arange(0, 13) / 12))

t_index = np.random.randint(len(f))
tonic_f = f[t_index]

while True:
  q_index = np.random.randint(len(f))
  if q_index != t_index:
    break
question_f = f[q_index]

A = 1
Fs = 48000
d_t = 1.0
t = np.linspace(0., d_t, int(Fs*d_t))

tonic = list(A * np.sin(2.0 * np.pi * tonic_f * t))
silence = list(A * np.sin(2.0 * np.pi * 0 * t))
question = list(A * np.sin(2.0 * np.pi * question_f * t))

y = tonic + silence + question

IPython.display.Audio(y, rate=Fs, autoplay=True)
```
