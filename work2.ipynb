```python
import numpy as np
import IPython

notes = ['ラ', 'ラ#', 'シ', 'ド', 'ド#', 'レ', 'レ#', 'ミ', 'ファ', 'ファ#', 'ソ', 'ソ#']

tone_range = '\u4E2D\u97F3\u57DF' #@param ['低音域', '中音域', '高音域'] {allow-input: false}

if tone_range == '低音域':
  bass = 110
elif tone_range == '中音域':
  bass = 220
elif tone_range == '高音域':
  bass = 440

f = list(bass * 2 ** (np.arange(0, 12) / 12))

t_index = np.random.randint(len(f))
t_index = 11
tonic_f = f[t_index]

notes = notes[t_index:] + notes[:t_index]
notes.append(notes[0])

print(notes)

while True:
  q_index = np.random.randint(len(f))
  if q_index != t_index:
    break
question_f = f[q_index]

print(q_index)

if q_index < t_index:
  notes.reverse()

print(notes)

A = 1
Fs = 48000
d_t = 1.0
t = np.linspace(0., d_t, int(Fs*d_t))

tonic = list(A * np.sin(2.0 * np.pi * tonic_f * t))
silence = list(A * np.sin(2.0 * np.pi * 0 * t))
question = list(A * np.sin(2.0 * np.pi * question_f * t))

y = tonic + silence + question

IPython.display.Audio(y, rate=Fs, autoplay=True)
```
