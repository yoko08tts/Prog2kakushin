#相対音感クイズ

プログラムを実行すると問題が出題されます

```python
import numpy as np
import IPython

key_tonic = 'C' #@param ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B', 'random'] {allow-input: false}
key_scale = 'major' #@param ['major', 'minor','chromatic'] {allow-input: false}

onmei = ['ド', 'ド#', 'レ', 'レ#', 'ミ', 'ファ', 'ファ#', 'ソ', 'ソ#', 'ラ', 'ラ#', 'シ']
notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']

A = 220

if key_scale == 'chromatic':
  scale = np.arange(0, 12)
else:
  if key_scale == 'major':
    scale = [0,2,2,1,2,2,2]
  elif key_scale == 'minor':
    scale = [0,2,1,2,2,1,2]
  scale = np.array(scale)
  scale = np.cumsum(scale)

answers = []
for a in scale:
  answers.append(onmei[a])
print('選択肢は', answers, 'です')

if key_tonic == 'random':
  t_index = np.random.randint(len(notes))
else:
  t_index = notes.index(key_tonic)

tonic_f = A * 2 ** (t_index / 12)

scale_f = list(tonic_f * 2 ** (scale / 12))

q_index = np.random.randint(len(scale_f))
#q_index = 0
question_f = scale_f[q_index]

r = np.random.random()
if r <= 0.5:
  question_f /= 2
else:
  if q_index == 0:
    question_f *= 2

A = 1
Fs = 48000
d_t = 1.0
t = np.linspace(0., d_t, int(Fs*d_t))

tonic = list(A * np.sin(2.0 * np.pi * tonic_f * t))
silence = list(A * np.sin(2.0 * np.pi * 0 * t))
question = list(A * np.sin(2.0 * np.pi * question_f * t))

y = tonic + silence + question

#y = []
#for f in scale_f:
#  tone = list(A * np.sin(2.0 * np.pi * f * t))
#  y = y + tone

IPython.display.Audio(y, rate=Fs, autoplay=True)
```

answer で答えを選択してプログラムを実行してください

```python
answer = '\u30DF' #@param ['ド', 'ド#', 'レ', 'レ#', 'ミ', 'ファ', 'ファ#', 'ソ', 'ソ#', 'ラ', 'ラ#', 'シ'] {allow-input: false}

if answer == answers[q_index]:
  print('正解')
else:
  print('不正解')

print('正解は' , answers[q_index] , 'です')
```
